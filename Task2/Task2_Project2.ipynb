```python
import pandas as pd                             # Here we have imported all the libraries which is required for completion of the Project2.
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Downloads/sales_data.csv")
df.head()           # This will give the 5 rows of the data set with all columns.
```

```python
print(df.shape)     # shape will give the no. of rows and columns from the  dataset.
df.info()           # info will provide all the information about the dataframe like data types and missing values.
df.describe()       #  describe will give summary of numeric data like mean,count,standard deviation,min,max....
```

```python
df.drop_duplicates(inplace=True)        # this will drop all the duplicated present in the dataset and inplace = true make the changes in the original dataset.
df.head()
```

```python
df.fillna(df.mean(numeric_only=True), inplace=True )  # This will fill mean of the numbers in the place of null values in all numeric data.
df.fillna("Hello_Data", inplace=True)      # this will fill placeholder in the place of all null non numeric values.
df.head()
```

```python
Trend = df.groupby("Date")["Sales"].sum() # Here I grouped the data of sales by Date for trend analysis.
Trend.plot(figsize=(10,5))  # here 10 is width and 5 is height of graph.
plt.title("Sales Trend Analysis")  # this will show the title for the graph.
plt.show()
```

```python
plt.scatter(df["Discount"],df["Profit"])   # here I have created a scatter graph where X-axis shows Discount and Y-axis shows Profit.
plt.title("Discount Vs Profit")          # This will give the title to the scatter.
plt.show()
```

```python
SalesRegion = df.groupby("Region")["Sales"].sum()   # Here I grouped the data of sales by region for visualising the sales distribution over region.
SalesRegion.plot(kind="bar")
plt.title("Sales by Region")
plt.show()
```

```python
df.groupby("Category")["Sales"].sum().plot(kind="pie")  # Here I grouped the data of sales by category for visualising the sales distribution.
plt.title("Sales by Category")
plt.show()
```

```python
############## PREDICTIVE MODELING #################

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

x = df[["Profit","Discount"]]       # Here the profit and discount columns will work as input where sales column act as output.
y = df["Sales"]

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)  # here 0.2 means it 20% testing and 80% training

model = LinearRegression()   # LinearRegression finds formulae.
model.fit(x_train, y_train)   # Here the model will learn and try to understand how profit and discount will affect the sales and finds the best value that will fit in the formulae which i screated by linear regression.
y_pred = model.predict(x_test)  # here the model will guess the sales using profit and discount.

print(y_pred)



```
