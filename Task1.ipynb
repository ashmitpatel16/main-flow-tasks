```python
import pandas as pd
df = pd.read_csv("Downloads/student/student-mat.csv",sep=";")
print(df.head())  #it gives only first 5 rows and all columns
print(df.isnull().sum())  #.is null() tells us which column having null values and .sum() counts the number of null value contained by a column
```

```python
print(df.dtypes)  #it describes about the data type of the columns
print(df.shape)  #it describes about the size of datasets
print(df.drop_duplicates())  #it removes all 
```

```python
rowAvg=df["G3"].mean()  #.mean() thiswill give the average score or you can say mean also of ny column
print(rowAvg)
```

```python
print((df["G3"]>15).sum())  # this will give us the number of student who has scored more than 15 marks in G3
```

```python
cor =df["studytime"].corr(df["G3"])
print (cor) # the output is 0.09781 which means it barely have a relation between them
```

```python
avgGrade= df.groupby("sex")["G3"].mean()
print (avgGrade.idxmax()) # this will give which gender has a higher average in G3 marks
```

```python
import matplotlib.pyplot as plt
plt.hist(df["G3"],bins=10,edgecolor = "black")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=df['studytime']
y=df['G3']
plt.scatter(x,y)
plt.xlabel("studytime")
plt.ylabel("G3")
plt.title("Scatter Plot between x and y")
plt.show()
```

```python
import matplotlib.pyplot as plt
categories = ['Male', 'Female']
average_scores= df.groupby("sex")["G3"].mean()
plt.bar(categories, average_scores, color=['blue', 'pink'])


plt.xlabel('Gender')
plt.ylabel('Average Score')
plt.title('Average Scores of Male and Female Students')

# Show plot
plt.show()
```
